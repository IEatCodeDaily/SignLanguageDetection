```python
import numpy as np
import cv2
import tensorflow as tf
import keras
from keras.preprocessing.image import ImageDataGenerator
```

```python
dict = ['A','B','C','D','E','I','L','O','U','Y']
model = keras.models.load_model("edgedetect.model")
```

```python
cam = cv2.VideoCapture(2)
while True:
    ret, frame = cam.read()
    if ret:
        frame = cv2.flip(frame, 1)
        ROI = frame[0:300, 170:470]

        gray_frame = cv2.cvtColor(ROI, cv2.COLOR_BGR2GRAY)
        gray_blur = cv2.GaussianBlur(gray_frame,(3,3),cv2.BORDER_DEFAULT)
        edge_frame = cv2.Canny(gray_blur,0,100)

        pred_frame = np.expand_dims(edge_frame, axis = 0)
        #pred_frame = pred_frame.astype('float32')/255

        pred = model.predict(pred_frame)
        predicted_classes = np.argmax(pred, axis=1)
        print(dict[predicted_classes[0]])

        cv2.rectangle(frame,(170,0),(470,300),(0,255,0),2) 
        cv2.putText(frame, dict[np.argmax(pred)], (170, 45), cv2.FONT_HERSHEY_SIMPLEX, 1, (0,0,255), 2) 
        cv2.imshow("Sign Detection", frame)
        
        cv2.imshow("ROI", edge_frame)

    if cv2.waitKey(10) & 0xFF == ord('q'):
        break
cam.release()
cv2.destroyAllWindows()
```
